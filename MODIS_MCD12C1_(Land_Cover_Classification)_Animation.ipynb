```python
import geemap.geemap as gm
import ee
ee.Authenticate()
ee.Initialize(project='bha5vya')
Map = gm.Map()
```

```python
india = ee.FeatureCollection("projects/bha5vya/assets/National_Boundary")
Map.addLayer(india, {}, "India Boundary")
Map
```

```python
roi = ee.Geometry.Polygon(
    [[[67.060547, 4.740675],
      [67.060547, 38.065392],
      [98.964844, 38.065392],
      [98.964844, 4.740675],
      [67.060547, 4.740675]]]
)
```

```python
dataset = ee.ImageCollection("MODIS/061/MCD12C1").select("Majority_Land_Cover_Type_1")
landcover_time_series = dataset.filterDate("2001-01-01", "2023-01-01")

vis_params = {
    "min": 0,
    "max": 16,
    "palette": [
        "1c0dff", "05450a", "086a10", "54a708", "78d203", "009900", "c6b044", "dcd159",
        "dade48", "fbff13", "b6ff05", "27ff87", "c24f44", "a5a5a5", "ff6d4c",
        "69fff8", "f9ffa4"
    ],
}
```

```python
rgbVis = landcover_time_series.map(lambda img: img.visualize(**vis_params).clip(india))

gif_params = {
    "region": roi,
    "dimensions": 600,
    "crs": "EPSG:3857",
    "framesPerSecond": 3,
    "format": "gif",
}

print("Land Cover Time Series Animation:", rgbVis.getVideoThumbURL(gif_params))

Map.addLayer(dataset.median(), vis_params, "Land Cover (Median)")
Map
```
